```python
import random 

suits = ('Hearts', 'Spades', 'Clubs', 'Diamonds')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values =  {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':11, 'Queen':12, 'King':13, 'Ace':14}
```

```python
class Card():
    
    def __init__(self,suit,rank):
        
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
    def __str__(self):
        return self.rank + ' of ' + self.suit       
```

```python
class Deck():
    
    
    
    def __init__(self):
        
        self.all_cards = []
        for suit in suits:
            for rank in ranks:
                self.all_cards.append(Card(suit,rank))
                
                
                
    def shuffle(self):
        
        random.shuffle(self.all_cards)
        
        
        
    def deal_one(self):
        
        return self.all_cards.pop()
```

```python
class Player():
    
    
    
    def __init__(self,name):
        
        self.name = name
        self.player_card = []
    
    
    
    def remove_one(self):
        
        return self.player_card.pop(-1)
    
    
    
    def add_one(self,new_cards):
        
        if type(new_cards) == type ([]):
            return self.player_card.extend(new_cards)
        else:
            return self.player_card.append(new_cards)
        
    
    
    def __str__(self):
        
        return f"{self.name} has {len(self.player_card)} cards"
     
```

```python
Amir = Player('Amir')
Pegah = Player('Pegah')
```

```python
# Game Logic

mydeck = Deck()

for i in range(26):
    Amir.add_one(mydeck.deal_one())
    Pegah.add_one(mydeck.deal_one())

game_on = True
game_round=1
```

```python
while game_on:
    
    print(f'Round {game_round}')
    game_round+=1
    
    
    if len(Amir.player_card) == 0:
        print ('Amir has lost, Pegah Won!')
        game_on = False
        break
        
    if len(Pegah.player_card) == 0:
        print ('Pegah has lost, Amir Won!')
        game_on = False
        break
      
    
    
    
    
    Amir_game_card = []
    Amir_game_card.append(Amir.remove_one())

    Pegah_game_card = []
    Pegah_game_card.append(Pegah.remove_one())
    
    
    at_war = True
    
    
    
    
    while at_war:
    

    
    
        if Amir_game_card[-1].value > Pegah_game_card[-1].value:
            Amir.add_one(Amir_game_card)
            Amir.add_one(Pegah_game_card)
            at_war = False
            break
            
            
        elif Amir_game_card[-1].value < Pegah_game_card[-1].value:
            Pegah.add_one(Amir_game_card)
            Pegah.add_one(Pegah_game_card)
            at_war = False
            break
            
        else:
            
            print('WAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAR')
            if len(Amir.player_card) <= 5:
                print("Amir is unable to declare war! Game Over at War")
                game_on = False
                break
                
            elif len(Pegah.player_card) <= 5:
                print("Pegah is unable to declare war! Game Over at War")
                game_on = False
                break
                
            else:
                for i in range(5):
                    Amir_game_card.append(Amir.remove_one())
                    Pegah_game_card.append(Pegah.remove_one())
```
